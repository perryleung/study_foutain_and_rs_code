```python
# _*_ coding=utf-8 _*_ 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys, os
from math import log, ceil
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
df_rsd = pd.read_csv('./LT/RSD_Fri_Dec__4_17_29_29_2020/avgs_rsd_k1000.csv')
df_switch = pd.read_csv('./LT/open_Fri_Dec_11_20_05_41_2020/avgs_open.csv')
df_rsd_FB = pd.read_csv('./LT_feedback/RSD_Sat_Dec__5_11_01_15_2020/feedback_avgs_rsd_k1000.csv')
df_switch_FB = pd.read_csv('./LT_feedback/open_Fri_Dec_11_20_23_35_2020/feedback_avgs_open.csv')
K = list(df_rsd['K'])
avgs_rsd = list(df_rsd['avgs'])
avgs_switch = list(df_switch['avgs'])
avgs_rsd_FB = list(df_rsd_FB['avgs'])
avgs_switch_FB = list(df_switch_FB['avgs'])
# plot
plt.plot(K, avgs_rsd, marker='*', label='RSD(0.03,0.05) K-avgs')
plt.plot(K, avgs_rsd_FB, marker='o', label='RSD_Feedback(0.03,0.05) K-avgs')
plt.plot(K, avgs_switch, marker='*', label='Switch(0.075) K-avgs')
plt.plot(K, avgs_switch_FB, marker='o', label='Switch_Feedback(0.075) K-avgs')
plt.xlabel('K')
plt.ylabel('avgs')
plt.grid()
plt.legend()
plt.savefig('./report/RSD_Switch_K_avgs_1.pdf')
plt.show()
```

```python
pro_rsd = list(df_rsd['pro'])
pro_rsd_FB = list(df_rsd_FB['pro'])
pro_switch = list(df_switch['pro'])
pro_switch_FB = list(df_switch_FB['pro'])
# plot
plt.plot(K, pro_rsd, marker='*', label='RSD(0.03,0.05) K-pros')
plt.plot(K, pro_rsd_FB, marker='o', label='RSD_Feedback(0.03,0.05) K-pros')
plt.plot(K, pro_switch, marker='*', label='Switch(0.075) K-pros')
plt.plot(K, pro_switch_FB, marker='o', label='Switch_Feedback(0.075) K-pros')
plt.xlabel('K')
plt.ylabel('pros')
plt.grid()
plt.legend()
plt.savefig('./report/RSD_Switch_K_pros.pdf')
plt.show()
```
