```python
# _*_ coding=utf-8 _*_ 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys, os
from math import log, ceil
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
df_rsd_1 = pd.read_csv('./LT_feedback/rsd_short_Fri_Dec_11_11_30_54_2020/feedback_avgs_rsd_short_1.csv')
df_solid_1_1 = pd.read_csv('./LT_feedback/solid_1_short_Fri_Dec_11_11_46_52_2020/feedback_avgs_solid_1_short_1.csv')
df_solid_2_1 = pd.read_csv('./LT_feedback/solid_2_short_Fri_Dec_11_12_36_48_2020/feedback_avgs_solid_2_short_1.csv')
df_poisson_1 = pd.read_csv('./LT_feedback/poisson_short_Fri_Dec_11_16_32_17_2020/feedback_avgs_poisson_short_1.csv')
df_open_1 = pd.read_csv('./LT_feedback/open_short_Fri_Dec_11_21_19_07_2020/feedback_avgs_short_1.csv')
K_1 = list(df_rsd_1['K'])
avgs_rsd_1 = list(df_rsd_1['avgs'])
avgs_solid_1_1 = list(df_solid_1_1['avgs'])
avgs_solid_2_1 = list(df_solid_2_1['avgs'])
avgs_poisson_1 = list(df_poisson_1['avgs'])
avgs_open_1 = list(df_open_1['avgs'])
# plot
plt.plot(K_1, avgs_rsd_1, marker='*', label='RSD-feedback (0.03,0.05)')
plt.plot(K_1, avgs_solid_1_1, marker='*', label='New1-feedback')
plt.plot(K_1, avgs_solid_2_1, marker='*', label='Fixed-feedback')
plt.plot(K_1, avgs_poisson_1, marker='*', label='New3-feedback')
plt.plot(K_1, avgs_open_1, marker='*', label='Switch-feedback (0.075)')
plt.xlabel('K')
plt.ylabel('avgs')
plt.grid()
plt.legend()
plt.savefig('./report/short_1_feedback_RSD_solid_MRSD_poisson_switch_K_avgs.pdf')
plt.show()
```

```python
pro_rsd_1 = list(df_rsd_1['pro'])
pro_solid_1_1 = list(df_solid_1_1['pro'])
pro_solid_2_1 = list(df_solid_2_1['pro'])
pro_poisson_1 = list(df_poisson_1['pro'])
pro_open_1 = list(df_open_1['pro'])
# plot
plt.plot(K_1, pro_rsd_1, marker='*', label='RSD-feedback (0.03,0.05)')
plt.plot(K_1, pro_solid_1_1, marker='*', label='New1-feedback')
plt.plot(K_1, pro_solid_2_1, marker='*', label='Fixed-feedback')
plt.plot(K_1, pro_poisson_1, marker='*', label='New3-feedback')
plt.plot(K_1, pro_open_1, marker='*', label='Switch-feedback (0.075)')
plt.xlabel('K')
plt.ylabel('pros')
plt.grid()
plt.legend()
plt.savefig('./report/short_1_feedback_RSD_solid_MRSD_poisson_switch_K_pros.pdf')
plt.show()
```
