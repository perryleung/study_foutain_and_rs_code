```python
df_LT = pd.read_csv('./LT/RSD_Fri_Dec__4_17_29_29_2020/avgs_Sat_Dec__5_11_00_25_2020.csv')
df_LT_FB = pd.read_csv('./LT_feedback/RSD_Sat_Dec__5_11_01_15_2020/feedback_avgs_Sun_Dec__6_02_27_15_2020.csv')
df_EW = pd.read_csv('./EW/RSD_Sun_Dec__6_09_04_57_2020/avgs_Mon_Dec__7_12_18_16_2020.csv')
df_EW_FB = pd.read_csv('./EW_feedback/RSD_Sat_Dec__5_23_04_45_2020/feedback_EW_avgs_Sun_Dec__6_08_41_52_2020.csv')
K = list(df_LT['K'])
avgs_LT = list(df_LT['avgs'])
#K_LT_FB = list(df_LT_FB['K'])
avgs_LT_FB = list(df_LT_FB['avgs'])
#K_EW = list(df_EW['K'])
avgs_EW = list(df_EW['avgs'])
#K_EW_FB = list(df_EW_FB['K'])
avgs_EW_FB = list(df_EW_FB['avgs'])
plt.plot(K, avgs_LT, marker='*', label='RSD_LT(0.03,0.05) K-avgs')
plt.plot(K, avgs_LT_FB, marker='o', label='RSD_LT_Feedback(0.03,0.05) K-avgs')
plt.plot(K, avgs_EW, marker='*', label='RSD_EW(0.03,0.05) K-avgs')
plt.plot(K, avgs_EW_FB, marker='o', label='RSD_EW_Feedback(0.03,0.05) K-avgs')
plt.xlabel('K')
plt.ylabel('avgs')
plt.grid()
plt.legend()
plt.savefig('./jupyter/RSD_K_avgs.pdf')
plt.show()
```

```python
pro_LT = list(df_LT['pro'])
pro_LT_FB = list(df_LT_FB['pro'])
pro_EW = list(df_EW['pro'])
pro_EW_FB = list(df_EW_FB['pro'])
plt.plot(K, pro_LT, marker='*', label='RSD_LT(0.03,0.05) K-pros')
plt.plot(K, pro_LT_FB, marker='o', label='RSD_LT_Feedback(0.03,0.05) K-pros')
plt.plot(K, pro_EW, marker='*', label='RSD_EW(0.03,0.05) K-pro')
plt.plot(K, pro_EW_FB, marker='o', label='RSD_EW_Feedback(0.03,0.05) K-pros')
plt.xlabel('K')
plt.ylabel('pros')
plt.grid()
plt.legend()
plt.savefig('./jupyter/RSD_K_pros.pdf')
plt.show()
```
