```python
# _*_ coding=utf-8 _*_ 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys, os
from math import log, ceil
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
df_rsd = pd.read_csv('./LT_feedback/RSD_Sat_Dec__5_11_01_15_2020/feedback_avgs_rsd.csv')
df_MRSD = pd.read_csv('./LT_feedback/MRSD_Wed_Dec__9_09_27_20_2020/feedback_avgs_MRSD.csv')
df_poisson = pd.read_csv('./LT_feedback/poisson_Fri_Dec_11_10_32_23_2020/feedback_avgs_poisson.csv')
df_solid_1 = pd.read_csv('./LT_feedback/solid_1_Thu_Dec_10_22_33_30_2020/feedback_avgs_solid_1.csv')
df_solid_2 = pd.read_csv('./LT_feedback/solid_2_Thu_Dec_10_20_55_13_2020/feedback_avgs_solid_2.csv')
K = list(df_rsd['K'])
avgs_rsd = list(df_rsd['avgs'])
avgs_solid_1 = list(df_solid_1['avgs'])
avgs_solid_2 = list(df_solid_2['avgs'])
avgs_MRSD = list(df_MRSD['avgs'])
avgs_poisson = list(df_poisson['avgs'])
# plot
plt.plot(K, avgs_rsd, marker='*', label='RSD-feedback (0.03,0.05)')
plt.plot(K, avgs_solid_1, marker='*', label='solid_1-feedback')
plt.plot(K, avgs_solid_2, marker='*', label='solid_2-feedback')
plt.plot(K, avgs_MRSD, marker='*', label='MRSD-feedback')
plt.plot(K, avgs_poisson, marker='*', label='poisson-feedback')
plt.xlabel('K')
plt.ylabel('avgs')
plt.grid()
plt.legend()
plt.savefig('./jupyter/feedback_RSD_solid_MRSD_poisson_K_avgs.pdf')
plt.show()
```

```python
pro_rsd = list(df_rsd['pro'])
pro_solid_1 = list(df_solid_1['pro'])
pro_solid_2 = list(df_solid_2['pro'])
pro_MRSD = list(df_MRSD['pro'])
pro_poisson = list(df_poisson['pro'])
# plot
plt.plot(K, pro_rsd, marker='*', label='RSD-feedback (0.03,0.05)')
plt.plot(K, pro_solid_1, marker='*', label='solid_1-feedback')
plt.plot(K, pro_solid_2, marker='*', label='solid_2-feedback')
plt.plot(K, pro_MRSD, marker='*', label='MRSD-feedback')
plt.plot(K, pro_poisson, marker='*', label='poisson-feedback')
plt.xlabel('K')
plt.ylabel('pros')
plt.grid()
plt.legend()
plt.savefig('./jupyter/feedback_RSD_solid_MRSD_poisson_K_pros.pdf')
plt.show()
```
