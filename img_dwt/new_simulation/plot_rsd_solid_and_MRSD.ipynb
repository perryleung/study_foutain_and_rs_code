```python
# _*_ coding=utf-8 _*_ 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys, os
from math import log, ceil
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
df_rsd = pd.read_csv('./LT/RSD_Fri_Dec__4_17_29_29_2020/avgs_rsd.csv')
df_solid_1 = pd.read_csv('./LT/solid_1_Thu_Dec_10_21_29_06_2020/avgs_solid_1.csv')
df_solid_2 = pd.read_csv('./LT/solid_2_Thu_Dec_10_19_37_16_2020/avgs_solid_2.csv')
df_MRSD = pd.read_csv('./LT/MRSD_Tue_Dec__8_22_39_21_2020/avgs_MRSD.csv')
df_poisson = pd.read_csv('./LT/poisson_Fri_Dec_11_10_13_47_2020/avgs_poisson.csv')
K = list(df_rsd['K'])
avgs_rsd = list(df_rsd['avgs'])
avgs_solid_1 = list(df_solid_1['avgs'])
avgs_solid_2 = list(df_solid_2['avgs'])
avgs_MRSD = list(df_MRSD['avgs'])
avgs_poisson = list(df_poisson['avgs'])
# plot
plt.plot(K, avgs_rsd, marker='*', label='RSD (0.03,0.05) K-avgs')
plt.plot(K, avgs_solid_1, marker='*', label='New1 K-avgs')
plt.plot(K, avgs_solid_2, marker='*', label='Fixed K-avgs')
plt.plot(K, avgs_MRSD, marker='*', label='New2 K-avgs')
plt.plot(K, avgs_poisson, marker='*', label='New3 K-avgs')
plt.xlabel('K')
plt.ylabel('avgs')
plt.grid()
plt.legend()
plt.savefig('./report/RSD_solid_MRSD_poisson_K_avgs.pdf')
plt.show()
```

```python
pro_rsd = list(df_rsd['pro'])
pro_solid_1 = list(df_solid_1['pro'])
pro_solid_2 = list(df_solid_2['pro'])
pro_MRSD = list(df_MRSD['pro'])
pro_poisson = list(df_poisson['pro'])
# plot
plt.plot(K, pro_rsd, marker='*', label='RSD(0.03,0.05) K-pro')
plt.plot(K, pro_solid_1, marker='*', label='New1 K-pro')
plt.plot(K, pro_solid_2, marker='*', label='Fixed K-pro')
plt.plot(K, pro_MRSD, marker='*', label='New2 K-pro')
plt.plot(K, pro_poisson, marker='*', label='New3 K-pro')
plt.xlabel('K')
plt.ylabel('pros')
plt.grid()
plt.legend()
plt.savefig('./report/RSD_solid_MRSD_poisson_K_pros.pdf')
plt.show()
```
