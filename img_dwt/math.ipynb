```python
# _*_ coding=utf-8 _*_ 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
```

```python
def com_drop(ser):
    per = 0
    for i in range(0, 17):
        per += comb(255, i) * (ser ** i) * ((1 - ser)**(255-i))
    return 1 - per

ber = np.arange(0.0001, 0.017, 0.001)
ser = [1 - (1-ii)**8 for ii in ber]
rs_per = [com_drop(ii) for ii in ser]
per = [1 - (1 - ii)**(8 * 255) for ii in ber]

plt.plot(ber, rs_per, marker='*', label='rs code')
plt.plot(ber, per, marker='+', label='without rs code')
# print(ber)
# print(per)
# print(rs_per)

plt.xlabel('BER')
plt.ylabel('PER')
plt.legend()
plt.grid()
plt.savefig('./simulation/sim_rs/per_rs.pdf')
plt.show()
```
